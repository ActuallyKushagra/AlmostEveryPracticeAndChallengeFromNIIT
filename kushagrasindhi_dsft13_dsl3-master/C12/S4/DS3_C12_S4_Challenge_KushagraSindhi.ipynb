```python
# task 1

import re
import spacy
from spacy.lang.en import English
nlp = spacy.load('en_core_web_sm')
import collections
from typing import Dict, List, Tuple
from spacy.matcher import Matcher, PhraseMatcher  
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
doc = open('DS3_C2_S4_GettysburgAddress_Data_Challenge.txt','r').read()
doc
```

```python
# task 2

def word2bow(words:List[str],dictionary : Dict[str,int])->List[Tuple[int,int]]:
    word_frequencies = collections.defaultdict(int)
    for word in words:
        if word not in dictionary:
            dictionary[word] = len(dictionary)
        word_frequencies[dictionary[word]] += 1
    return list(word_frequencies.items())

dictionary = {}
print('\n Bag of Words representation : \n',word2bow(doc.split(),dictionary))
print('\nDictionary \n', dictionary)
```

```python
# task 3

matcher = PhraseMatcher(nlp.vocab)
input_pat = ['battlefeild','battle-field']
pattern = [nlp(i) for i in input_pat]
matcher.add('BattleField',pattern)
doc = nlp(doc)
k = matcher(doc)
print(k)

matcher1 = Matcher(nlp.vocab)
pattern_1 = [{'LOWER':'battlefeild'}]
pattern_2 = [{'LOWER':'battle'},{'IS_PUNCT' : True},{'LOWER':'field'}]
matcher1.add('Task4a',[pattern_1,pattern_2])

k = matcher1(doc)
print(k)
```

```python
# task 5

matcher3 = PhraseMatcher(nlp.vocab)
pat = ['long endure', 'resting place']
pattern2 = [nlp(i) for i in pat]
matcher3.add('Task5',pattern2)

m = matcher3(doc)
print(m)
for match_id, start, end in m:
    span = doc[start:end]
    print(span.text, start, end)
```

```python
# task 7

tok1 = []
tok2 = []
sim = []

for token1 in doc:
    for token2 in doc:
        tok1.append(token1.text)
        tok2.append(token2.text)
        sim.append(token1.similarity(token2))
```
