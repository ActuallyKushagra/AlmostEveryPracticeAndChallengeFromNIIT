```python
import spacy
import re
from collections import Counter
```

```python
nlp = spacy.load('en_core_web_sm')
nlp.disable_pipes('tagger', 'parser', 'lemmatizer', 'ner')
```

```python
# Task 1

f1 = open('DS3_C2_S2_48hrs_Data_Practice.txt','r').read()
f2 = open('DS3_C2_S2_Scifiscripts_Data_Practice.txt', 'r').read()
f3 = open('DS3_C2_S2_TenThingsIHate_Data_Practice.txt','r').read()
```

```python
f1 = nlp(f1)
f2 = nlp(f2)
f3 = nlp(f3)
```

```python
count48 = Counter()
```

```python
for token in f1:
    count48[token.orth_]+=1
```

```python
# task 4

countthings = Counter()
countscifi = Counter()

for token in f2:
    countthings[token.orth_]+=1
    
for token in f3:
    countscifi[token.orth_]+=1
```

```python
thecount48 = dict(count48).get('the') + dict(count48).get('The')
thecountscifi = dict(countscifi).get('the') + dict(countscifi).get('The')
thecountthings = dict(countthings).get('the') + dict(countthings).get('The')
```

```python
print(f"""Total number of 'The' and 'the' in 48hrs.txt are = {thecount48}
Total number of 'The' and 'the' in scifiscripts.txt are = {thecountscifi}
Total number of 'The' and 'the' in tenthingsihate.txt are = {thecountthings}""")
```

As we can see, 'Ten Things I hate' have the least number of The's and the's at just 664
