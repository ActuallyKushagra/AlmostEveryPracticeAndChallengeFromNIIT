```python
from scipy.stats import norm, chi2, chi2_contingency
import numpy as np
import matplotlib.pyplot as plt
```

###### Task 2



###### 

Ho: Early and regular applications have same distribution

Ha: Early and regular applications have different distribution

```python
observed_data = [[2115, 1792, 5304, 1734, 2685, 1245], [577, 627, 1228, 444, 1195, 761]]
observed_data
```

```python
alpha = 0.05
```

```python
statistics, pval, dof, expected = chi2_contingency(observed_data)
```

```python
print(f"""Statistics = {statistics}
Pvalue = {pval}
dof = {dof}
expected = {expected}""")
```

```python
crit = chi2.ppf(1-alpha , dof)
crit
```

```python
if pval > alpha and statistics < crit:
    print('Null hypothesis accepted')
else:
    print('reject null hypothesis')
```

```python
x = np.arange(40)

plt.plot(x, chi2.pdf(x, dof), label = 'df')
plt.plot(x, chi2.pdf(x, dof-1), label = 'df-1')
plt.plot(x, chi2.pdf(x, dof-2), label = 'df-2')
plt.plot(x, chi2.pdf(x, dof-3), label = 'df-3')
plt.plot(x, chi2.pdf(x, dof-4), label = 'df-4')
plt.legend()
```

###### Task 2



###### 

Ho: Males and Females feel the same way

Ha: Males and Females feel differently

```python
alpha = 0.05
observed_data = [42, 95, 27, 63]
n = len(observed_data)
expected = sum(observed_data)/n
dof = n-1
```

```python
crit = chi2.ppf(1-alpha, dof)
statistics = 0
for i in observed_data:
    statistics+=(i-expected)**2/expected
pval = norm.sf(abs(statistics), dof)
print(f"""Pvalue = {pval}
Statistics = {statistics}
Critical = {crit}
dof = {dof}
n = {n}
alpha = {alpha}""")
```

```python
if pval > alpha and statistics < crit:
    print('Null hypothesis accepted')
else:
    print('reject null hypothesis')
```

```python
x = np.arange(40)

plt.plot(x, chi2.pdf(x, dof), label = 'df')
plt.plot(x, chi2.pdf(x, dof-1), label = 'df-1')
plt.plot(x, chi2.pdf(x, dof-2), label = 'df-2')
plt.legend()
```
