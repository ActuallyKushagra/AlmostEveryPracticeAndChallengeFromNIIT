```python
# b

def word2bow(words:List[str],dictionary : Dict[str,int])->List[Tuple[int,int]]:
    word_frequencies = collections.defaultdict(int)
    for word in words:
        if word not in dictionary:
            dictionary[word] = len(dictionary)
        word_frequencies[dictionary[word]] += 1
    return list(word_frequencies.items())

dictionary = {}
print ('Sample Sentence \n', txt3)
print('\n Bag of Words representation : \n',word2bow(txt3.split(),dictionary))
print('\nDictionary \n', dictionary)
```

```python
# task 2

t2txt = nlp('Hey, Siri! Hey, Siri!')

pattern = [{'LOWER': 'hey'},{'IS_PUNCT':True},{'LOWER':'siri'}]
pattern2 =  [{'LOWER':'hey'},{'LOWER':'siri'}]
matcher.add("HeySiri!",[pattern,pattern2])
matches=matcher(t2txt)
matches
```

```python
matcher.add("HelloWorld",[pattern2])
doc = nlp('Hey, Siri! Hey, Siri!')
matches2=matcher(doc)
for match_id,start,end in matches:
    string_id = nlp.vocab.strings[match_id]
    span = doc[start:end]
    print(span.text)
```

```python
# task 4

matcher2 = PhraseMatcher(nlp.vocab)
lst = ['ROTTEN mangoes', 'sweet oranges']
pattern= [nlp(i) for i in lst]
matcher2.add("RottenMangoesSweetOranges",pattern)

doc = nlp("Do not put rotten mangoes and sweet oranges together")
k = matcher2(doc)
print(k)
for match_id, start, end in k:
    span = doc[start:end]
    print(span)
```
