```python
import pandas as pd
import spacy
import re
nlp = spacy.load("en_core_web_sm")
import nltk
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
nltkstop = stopwords.words('english') 
```

```python
# task 1

df = pd.read_csv('DS3_C2_S3_CookingReview_Data_Challenge.csv')
df
```

```python
# task 2

para = " ".join(i for i in df["content"])
para[:1000]
```

```python
filtered = []

for word in nlp(para[:1000000]):
    if word.is_stop == False:
        filtered.append(word)

filtered
```

```python
# task 3

rtext = df.iloc[0]['content']
rtext
```

```python
# stemming
stemmer = SnowballStemmer(language='english')

li = list(rtext.split(" "))
nwords = " "
for token in li:
    print(token, '--->',stemmer.stem(token))
    nwords = nwords + ' ' + stemmer.stem(token)
    
nwords
```

```python
# lemmatization
nword = " "
for word in nlp(rtext):
    print(word.text,'--->',word.lemma_)
    nword = nword + " " + word.lemma_
```

```python
# task 4
#a
word = []
pos = []
tag = []
for j in nlp(rtext):
    print('The word is',j,'\t\tand its POS is=>',j.pos_,'\t\t and its tag is :',j.tag_)
    word.append(j)
    pos.append(j.pos_)
    tag.append(j.tag_)
```

```python
#b
prop = pd.DataFrame({'word':word, 'pos': pos, 'tag': tag})
prop
```

```python
prop.groupby(['pos']).count()
```
